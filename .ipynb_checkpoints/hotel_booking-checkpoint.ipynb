# Importing the libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import jaccard_score, f1_score, accuracy_score
from sklearn.model_selection import train_test_split
```

# Read the DataFrame.

```python
df = pd.read_csv("./hotel_bookings.csv")
```

# Get some info about the DataFrame.

```python
df.info()
```

```python
df.tail().T
```

```python
df.isna().sum()
```

# Working with NaN.

```python
df.drop("company", axis=1, inplace=True)
```

```python
df["children"].fillna(df["children"].mode()[0], inplace=True)
```

```python
df["country"].fillna(df["country"].mode()[0], inplace=True)
```

```python
df["agent"].fillna(df["agent"].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

# Working with object values

```python
target = []
for i in df.columns:
    col = df[i]
    if type(col[0]) == str:
        target.append(i)
print(target)
for i in target:
    encoder = LabelEncoder()
    encoder.fit(list(df[i]))
    df[i] = encoder.fit_transform(df[i])
df.head().T
```

# Features selection

```python
x = df.drop("is_canceled", axis = 1)
y = df["is_canceled"]
```

# Data splitting.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

# LogisticRegression.

```python
LR = LogisticRegression().fit(x_train, y_train)
```

```python
y_predLR = LR.predict(x_test)
accLR = accuracy_score(y_test, y_predLR)
accLR
```

# KNN.

```python
KNN = KNeighborsClassifier(n_neighbors=5).fit(x_train, y_train)
```

```python
y_predKNN = KNN.predict(x_test)
accKNN = accuracy_score(y_test, y_predKNN)
accKNN
```

# RandomForest.

```python
RF = RandomForestClassifier(max_depth=15, n_jobs=-1).fit(x_train, y_train)
```

```python
y_predRF= RF.predict(x_test)
accRF = accuracy_score(y_test, y_predRF)
accRF
```

# DecisionTree.

```python
DT = DecisionTreeClassifier(max_depth=25).fit(x_train, y_train)
```

```python
y_predDT= DT.predict(x_test)
accDT = accuracy_score(y_test, y_predDT)
accDT
```

# All models acc.

```python
allacc = pd.DataFrame(columns=["Accuracy"], index=["KNN", "Random Forest",
                                                  "Decision Tree", "Logistic Regression"])
models = {'KNN' : KNN, 'Random Forest' : RF, 'Decision Tree' : DT, "Logistic Regression" : LR}

for name, model in models.items() :
  allacc.loc[name, 'Accuracy'] = accuracy_score(y_test, y_pred = model.predict(x_test))
```

```python
allacc
```

```python
allacc.unstack().plot(kind = 'bar', stacked = True, figsize = (12, 7), title = 'Final models accuracy')
```
